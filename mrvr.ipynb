```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
import holoviews as hv
```

```python
df = pd.read_csv(Path('historical_data_bitcoin_20220606114445.csv'))
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df.head()
```

```python
df = df.set_index('Date')
df.head()
```

```python
df['MVRV'] = df['Circulating Marketcap'] / df['Realized Marketcap']
df.head()
```

```python
first_plot = df.hvplot(y = ['Circulating Marketcap', 'Realized Marketcap'])
```

```python
mvrv_plot = df.hvplot(y = 'MVRV')
bottom = hv.HLine(1)
top = hv.HLine(3.7)
```

```python
(mvrv_plot * bottom * top) + first_plot
```
